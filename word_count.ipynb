```python
#coding=utf-8
#其实以前写过这样的东西来着= =……
import re,sys,os

Fliter = re.compile("[^A-Za-z-\']|((?<![A-Za-z])[-\'])|([-\'](?![A-Za-z]))")
Divider = re.compile("\s")
File = open('test.txt').read()
Data =  Divider.split(Fliter.sub(" ",File))
Dict = {}
for i in Data:
	j = i.lower()
	try:
		Dict[j]+=1
	except KeyError:
		Dict[j] =1
	except:
		raise 
for i in  Dict.items():
	print "%s:%s"%i
```

```python
#-*- coding:utf-8-*-

with open('test.txt', 'r') as f:
    from collections import Counter
    c = Counter()
    for line in f:
        words = line.split()
        #print(words)
        c += Counter(words)
    for words in c.most_common():
        print(words)
```
