```python
import requests
res = requests.get("https://www.youtube.com/watch?v=izB8P8oyhJk")
print res.text
```

```python
import re
m = re.search('"args":({.*?}),',res.text)
```

```python
print m.group(1)
```

```python
import json
jd=json.loads(m.group(1))
print jd['url_encoded_fmt_stream_map']
```

```python
import urlparse
a=urlparse.parse_qs(jd['url_encoded_fmt_stream_map'])
print a['url']
```

```python
print a['url'][0]
import shutil
res2=requests.get(a['url'][0],stream=True)
f=open('b.mp4','wb')
shutil.copyfileobj(res2.raw,f)
f.close()
```
